```python
from python_files.import_libs import *
from python_files.model import *
```

```python
# Load datas
model_df = pd.read_csv('data/model_df.csv')
arima_df = pd.read_csv('data/arima_df.csv').set_index('date')
```

```python
# drop the columns sales and date
data = model_df.drop(['sales','date'],axis=1)
# absorving the 12 last values for train and test data
train, test = data[0:-12].values, data[-12:].values
```

```python
X_train, y_train, scaler = scale_data_train(train)
X_test, y_test = scale_data_test(test, scaler)
```

```python
# Linear Regression
run_model(train, test, LinearRegression(), 'LinearRegression')
```

```python
# Randon forest regressor
run_model(train, test, RandomForestRegressor(n_estimators=100, max_depth=20), 'RandomForest')
```

```python
# XGBoost
run_model(train, test, XGBRegressor( n_estimators=100, 
                                    learning_rate=0.2, 
                                    objective='reg:squarederror'), 'XGBoost')
```

```python
# LSTM
# Refference https://towardsdatascience.com/5-machine-learning-techniques-for-sales-forecasting-598e4984b109#:~:text=Forecasting%20sales%20is%20a%20common,demand%2C%20and%20project%20future%20budgets.
# Refference https://keras.io/api/layers/recurrent_layers/lstm/
# Refference https://keras.io/guides/sequential_model/
# Refference https://keras.io/api/layers/core_layers/dense/
# Refference https://keras.io/api/models/model_training_apis/
# Set the train and test data wtih 
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
# Applying Sequential class to the model, as it has a simple stack of layers and each layer has exactly one input and one output tensor
model = Sequential()
# Add the LSTM layer to the model. The LSTM layer receives:
# units: output space dimension 4
# batch_input_shape: batch size 1, time step equal to training data index 1 shape and data dimension equal to training data index 2 shape
# stateful: True to use the last state of each sample at index i in one batch being used as the initial state for the sample at index i in the next batch
model.add(LSTM(4, batch_input_shape=(1, X_train.shape[1], X_train.shape[2]), stateful=True))
# Adding Dense to the model with dimensionality 1
model.add(Dense(1))
model.add(Dense(1))
# Setting up the model for training
# loss: Instance of tf.keras.losses.Loss that returns a loss float tensor based on root mean square error
# optimizer: Optimizer Adam
model.compile(loss='mean_squared_error', optimizer='adam')
# Training the model for a fixed number of epochs with the fit
# X_trian: input data
# y_train: target data
# epochs: 200 times to train the model
# bathc_size 1 swatch per gradient update
# verbose: 1 for progress bar and verbosity
# shffle: Specified as false to not shuffle training data every epoch
model.fit(X_train, y_train, epochs=200, batch_size=1, verbose=1, shuffle=False)

predictions = model.predict(X_test,batch_size=1)
    
original_df = load_original()
unscaled = scaling(predictions, X_test, scaler, lstm=True)
unscaled_df = predict_data(unscaled, original_df)
    
get_scores(original_df, 'LSTM', unscaled_df)
    
plot_results(unscaled_df, original_df, 'LSTM')
```
